```python
import pandas as pd 
import numpy as np
from unidecode import unidecode
```

```python
file_path = "Resources/space_missions.csv"
encoding =  "ISO-8859-1"
space_missions_df = pd.read_csv(file_path, encoding=encoding)
```

```python
space_missions_df.head()
```

```python
space_missions_df["Mission"] = space_missions_df["Mission"].apply(unidecode)
space_missions_df.head()
```

```python
unique_launch_sites = space_missions_df['Location'].unique()
unique_launch_sites
```

```python
del space_missions_df["Price"]
space_missions_df.head()
```

```python
space_missions_df["Time"] = space_missions_df["Time"].astype(str)
```

```python
day_night = []
```

```python
for row in space_missions_df["Time"]:
    first_two = row[:2]
    
    try:
        
        if int(first_two) >= 6 and int(first_two) < 18:
            day_night.append("Day")
        else: 
            day_night.append("Night")
    except ValueError:
        day_night.append("NA")
        
```

```python
space_missions_df["Day/Night Launch"] = day_night
space_missions_df.head()
```

```python
#Cleanup of the Locations to add values or null values bef turning it into a new CSV file
#Starting with the Uchinoura Space Center, to which there are only two locations. One needs an empty space first. The other needs no
#alteration, as it already contains the value of 'Mu Pad' as the value needed.
def fix_location(location, center, excluded_rows=None):
    if excluded_rows is None:
        excluded_rows = []
    if center in location and 'Mu Pad' not in location and location not in excluded_rows:
        return f' , {center}'
    return location
space_missions_df['Location'] = space_missions_df.apply(lambda row: fix_location(row['Location'], 'Uchinoura Space Center, Japan'), axis=1)
#-----------------------------------------------------
#Exclude for the Taiyuan Satellite Launch Center
excluded_rows_taiyuan = [
    'LC-7, Taiyuan Satellite Launch Center, China',
    'LC-9, Taiyuan Satellite Launch Center, China',
    'LC-16, Taiyuan Satellite Launch Center, China',
    'LC-9A, Taiyuan Satellite Launch Center, China']
space_missions_df['Location'] = space_missions_df.apply(lambda row: fix_location(row['Location'], 'Taiyuan Satellite Launch Center, China', excluded_rows_taiyuan), axis=1)
#-----------------------------------------------------
#Exclude for the Jiuquan Satellite Launch Center
excluded_rows_jiuquan = [
    'Site 5020 (LA-2A), Jiuquan Satellite Launch Center, China',
    'Site 138 (LA-2B), Jiuquan Satellite Launch Center, China',
    'Site 901 (SLS-1), Jiuquan Satellite Launch Center, China',
    'Site 9401 (SLS-2), Jiuquan Satellite Launch Center, China',
    'Site 95, Jiuquan Satellite Launch Center, China']
space_missions_df['Location'] = space_missions_df.apply(lambda row: fix_location(row['Location'], 'Jiuquan Satellite Launch Center, China', excluded_rows_jiuquan), axis=1)
#-----------------------------------------------------
#Exclude for the Xichang Satellite Launch Center
excluded_rows_xichang = [
    'LC-3, Xichang Satellite Launch Center, China',
    'LC-2, Xichang Satellite Launch Center, China']
space_missions_df['Location'] = space_missions_df.apply(lambda row: fix_location(row['Location'], 'Xichang Satellite Launch Center, China', excluded_rows_xichang), axis=1)
#----------------------------------------------------
#Just use the str.replace for the rest of the values.
#Change Svobodny Cosmodrome, Russia, to have a space before Svobodny...
space_missions_df['Location'] = space_missions_df['Location'].str.replace('Svobodny Cosmodrome, Russia', ' , Svobodny Cosmodrome, Russia', case=False)
# Changing Hammaguir, Algeria to Hammaguir-Algeria
space_missions_df['Location'] = space_missions_df['Location'].str.replace('Brigitte, Hammaguir, Algeria, France', 'Brigitte, Hammaguir - Algeria, France', case=False)
#Need to change Guiana Space Centre, French Guiana to Guiana Space Cenre-French Guiana
space_missions_df['Location'] = space_missions_df['Location'].str.replace('ELD, Guiana Space Centre, French Guiana, France', 'ELD, Guiana Space Centre - French Guiana, France', case=False)
space_missions_df['Location'] = space_missions_df['Location'].str.replace('ELA-1, Guiana Space Centre, French Guiana, France', 'ELA-1, Guiana Space Centre - French Guiana, France', case=False)
space_missions_df['Location'] = space_missions_df['Location'].str.replace('ELV-1 (SLV), Guiana Space Centre, French Guiana, France', 'ELV-1 (SLV), Guiana Space Centre - French Guiana, France', case=False)
space_missions_df['Location'] = space_missions_df['Location'].str.replace('ELA-2, Guiana Space Centre, French Guiana, France', 'ELA-2, Guiana Space Centre - French Guiana, France', case=False)
space_missions_df['Location'] = space_missions_df['Location'].str.replace('ELA-3, Guiana Space Centre, French Guiana, France', 'ELA-3, Guiana Space Centre - French Guiana, France', case=False)
space_missions_df['Location'] = space_missions_df['Location'].str.replace('ELS, Guiana Space Centre, French Guiana, France', 'ELS, Guiana Space Centre - French Guiana, France', case=False)
space_missions_df['Location'] = space_missions_df['Location'].str.replace('ELV-1, Guiana Space Centre, French Guiana, France', 'ELV-1, Guiana Space Centre - French Guiana, France', case=False)
#Need to change Alcântara Launch Center, Maranhão to Alcântara Launch Center - Maranhão
space_missions_df['Location'] = space_missions_df['Location'].str.replace('VLS Pad, Alcântara Launch Center, Maranhão, Brazil', 'VLS Pad, Alcântara Launch Center - Maranhão, Brazil', case=False)
#Need to change Pacific Spaceport Complex, Kodiak to Pacific Spaceport Complex - Kodiak
space_missions_df['Location'] = space_missions_df['Location'].str.replace('LP-3B, Pacific Spaceport Complex, Kodiak, Alaska, USA', 'LP-3B, Pacific Spaceport Complex - Kodiak, Alaska, USA', case=False)
#Need to add USA after New Mexico
space_missions_df['Location'] = space_missions_df['Location'].str.replace('Vertical Launch Area, Spaceport America, New Mexico', 'Vertical Launch Area, Spaceport America, New Mexico, USA', case=False)
#Need to add China after Yellow Sea
space_missions_df['Location'] = space_missions_df['Location'].str.replace('Tai Rui Barge, Yellow Sea', 'Tai Rui Barge, Yellow Sea, China', case=False)
space_missions_df['Location'] = space_missions_df['Location'].str.replace('DeBo 3 Barge, Yellow Sea', 'DeBo 3 Barge, Yellow Sea, China', case=False)
#Need to fix typo in Plateform to Platfomr and add Iran at the end.
space_missions_df['Location'] = space_missions_df['Location'].str.replace('Launch Plateform, Shahrud Missile Test Site', 'Launch Platform, Shahrud Missile Test Site, Iran', case=False)
#Fixing LP-41, Kauai, Pacific Missile Range Facility to LP-41, Kauai Pacific Missile Range Facility, Hawaii, USA
space_missions_df['Location'] = space_missions_df['Location'].str.replace('LP-41, Kauai, Pacific Missile Range Facility', 'LP-41, Kauai Pacific Missile Range Facility, Hawaii, USA', case=False)
#Fixing Rocket Lab LC-1B, MÄ\x81hia Peninsula, New Zealand to: Rocket Lab LC-1B, Māhia Peninsula, New Zealand
space_missions_df['Location'] = space_missions_df['Location'].str.replace('Rocket Lab LC-1B, MÄ\x81hia Peninsula, New Zealand', 'Rocket Lab LC-1B, Mahia Peninsula, New Zealand', case=False)
#Fixing Rocket Lab LC-1A, MÄ\x81hia Peninsula, New Zealand, to: Rocket Lab LC-1A, Māhia Peninsula, New Zealand
space_missions_df['Location'] = space_missions_df['Location'].str.replace('Rocket Lab LC-1A, MÄ\x81hia Peninsula, New Zealand', 'Rocket Lab LC-1A, Mahia Peninsula, New Zealand', case=False)
space_missions_df['Location'] = space_missions_df['Location'].str.replace('Rocket Lab LC-1A, M?hia Peninsula, New Zealand', 'Rocket Lab LC-1A, Mahia Peninsula, New Zealand', case=False)
```
