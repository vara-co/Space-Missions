```python
import pandas as pd 
import numpy as np
from unidecode import unidecode
import pycountry #in order to import this and use it, you first need to open Terminal or Gitbash, activate
                    #your dev environment using conda activate dev and then run the command pip install pycountry
```

```python
file_path = "Resources/space_missions.csv"
encoding =  "ISO-8859-1"
space_missions_df = pd.read_csv(file_path, encoding=encoding)
```

```python
space_missions_df.head()
```

```python
space_missions_df["Mission"] = space_missions_df["Mission"].apply(unidecode)
space_missions_df.head()
```

```python
del space_missions_df["Price"]
space_missions_df.head()
```

```python
space_missions_df["Time"] = space_missions_df["Time"].astype(str)
```

```python
day_night = []
```

```python
for row in space_missions_df["Time"]:
    first_two = row[:2]
    
    try:
        
        if int(first_two) >= 6 and int(first_two) < 18:
            day_night.append("Day")
        else: 
            day_night.append("Night")
    except ValueError:
        day_night.append("NA")
        
```

```python
space_missions_df["Day/Night Launch"] = day_night
space_missions_df.head()
```

```python
# create a new column filled with nulls
space_missions_df["Country"] = np.nan
space_missions_df.head()
```

```python
#search for country names and write them into the country column
i = 0 #necessary to set i as 0 so the iloc runs properly. This code only worked with the iloc method
for row in space_missions_df["Location"]:
    text = space_missions_df.iloc[i,1] #sets text as the value of current row, column 1
    for country in pycountry.countries: #loops through the pycountry library and searches each row for existing country names
        if country.name in text:
            country_name = country.name
            space_missions_df.iloc[i,9] = country_name #overwrites the null value with the country name
    i += 1 #sets i as the value of the next row for next iloc
space_missions_df.head()
```

```python
# "USA" doesn't exist in the pycountry library (It's probably "United States")
#also had to do this as a separate loop because combining it with above loop caused errors for unkown reasons
i = 0 # same method as above
for row in space_missions_df["Location"]:
    if "USA" in space_missions_df.iloc[i,1]: #searches for "USA" within the string
        country_name = "USA"
        space_missions_df.iloc[i,9] = country_name #overwrites nulls with USA
    i += 1
space_missions_df.head()
```

```python
space_missions_df["Country"].value_counts() #Yay! Here's all our countries
```
